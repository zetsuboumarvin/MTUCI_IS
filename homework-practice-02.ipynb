```python
def pairplot(df, target):
    ncol, nrow = 7, df.shape[1] // 7 + (df.shape[1] % 7 > 0)
    plt.figure(figsize=(ncol * 4, nrow * 4))

    for i, feature in enumerate(df.columns):
        plt.subplot(nrow, ncol, i + 1)
        plt.scatter(df[feature], target, s=10, marker='o', alpha=.6)
        plt.xlabel(feature)
        if i % ncol == 0:
            plt.ylabel('target')
```

В данном задании мы рассмотрим стандартный датасет для задачи регрессии Boston Housing, в котором необходимо предсказать стоимость недвижимости по 13 признакам.

Датасет достаточно просто загрузить из библиотеки scikit-learn:

```python
data = datasets.load_boston()
df = pd.DataFrame(data.data)
df.columns = data.feature_names
y = data.target
```

Разделим выборку на обучающую и тестовую в отношении 8/2:

__(1 балл)__

Зачастую признаки в сырых данных имеют разный масштаб. Попробуйте применить масштабирование к данным, сравните качество. Заметно ли изменение?

Pipeline, StandardScaler, MinMaxScaler

```python
alphas = np.logspace(-6, 6, 100)


# Lasso
#  StandardScaler

pipe = Pipeline([('scaler', StandardScaler()), ('estimator', LassoCV(alphas = alphas))])
pipe.fit(x_train, y_train)
print("LassoCV and StandardScaler score = ", pipe.score(x_test, y_test))

# MinMaxScaler

pipe = Pipeline([('scaler', MinMaxScaler()), ('estimator', LassoCV(alphas = alphas))])
pipe.fit(x_train, y_train)
print("LassoCV and MinMaxScaler score = ", pipe.score(x_test, y_test))

# Ridge
#  StandardScaler

pipe = Pipeline([('scaler', StandardScaler()), ('estimator', RidgeCV(alphas = alphas))])
pipe.fit(x_train, y_train)
print("RidgeCV and StandardScaler score = ", pipe.score(x_test, y_test))

# MinMaxScaler

pipe = Pipeline([('scaler', MinMaxScaler()), ('estimator', RidgeCV(alphas = alphas))])
pipe.fit(x_train, y_train)
print("RidgeCV and MinMaxScaler score = ", pipe.score(x_test, y_test))
```
